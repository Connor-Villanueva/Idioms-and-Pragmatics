```python
import spacy
from spacy.matcher import PhraseMatcher
import json
import duckdb as db
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
DATA_PATH = "../Data/idiom_repository_all.json"
idiom_df = db.query(f"""
    SELECT *, list_concat(variations, [idiom]) AS all_variations
    FROM read_json_auto('{DATA_PATH}')
""").df()
```

```python
matcher = PhraseMatcher(nlp.vocab, attr="LOWER")
patterns = [nlp.make_doc(p) for q in idiom_df['all_variations'] for p in q]
matcher.add("PHRASES", patterns)
```

```python
doc = nlp("That guy was so gone, he was two sheets to the wind.")
```

```python
matches = matcher(doc)
for match_id, start, end in matches:
    span = doc[start:end]
    print(span.text, start, end)
```
