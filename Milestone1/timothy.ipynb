It appears that the variations column is mainly just for the variation of having 'a' or 'the' before a noun grouping (noun phrase?). There also appears to be some that are the difference of past, present, or future tense. Only two entries have more than 1 variation, id 2407 and id 5258, which are 'four sheets to the wind' and 'pot calling the kettle black', respectively.

```python
import duckdb as db

FILE_PATH = '../Data/idiom_repository_all.json'

idiom_df = db.query(f"""
    SELECT *
    FROM read_json_auto('{FILE_PATH}')
""").df()
```

```python
db.query(
"""
SELECT * 
FROM idiom_df
WHERE len(variations) > 0
         """)
```

```python
db.query(
"""
SELECT * 
FROM idiom_df
WHERE len(variations) > 1
         """)
```
